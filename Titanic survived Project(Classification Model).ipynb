#Now we have got the columns that are useful in predicting the survival of the passengers:

'Pclass': This column indicates the passenger class (1st, 2nd, or 3rd) and can be a good indicator of socioeconomic status. In general, passengers in higher classes may have had a better chance of survival due to their access to lifeboats and other resources.

'Sex': This column indicates the gender of the passenger and may also be a good predictor of survival, as women and children were given priority for evacuation during the disaster.

'Age': This column indicates the age of the passenger and could also be a good predictor of survival, as young children and elderly passengers may have had a harder time evacuating the ship or surviving the elements.

'SibSp' and 'Parch': These columns indicate the number of siblings/spouses and parents/children aboard the Titanic, respectively. These variables may be useful in predicting survival, as passengers traveling with family members may have had a greater likelihood of survival due to help from their companions.

'Embarked': This column indicates the port of embarkation for the passenger and could potentially be useful in predicting survival if the port is related to other factors that could affect a passenger's survival rate.

'Fare': This column could potentially be useful in predicting survival, as passengers who paid higher fares may have had access to better accommodations and resources on the ship, which could have increased their chances of survival.

#We can see outliers are present in 'Age','Fare'and 'FamilySize'

```python
#Removing outliers by Z score

from scipy.stats import zscore
out_features=df[['Age','Fare','FamilySize']]
z=np.abs(zscore(out_features))
z
```

```python
#threshold=3

np.where(z>3)
```

```python
#Now lets check with IQR method:

#1st Quantile
Q1=out_features.quantile(0.25)

#3rd Quantile
Q3=out_features.quantile(0.75)

IQR=Q3-Q1

df2=df[~((df < (Q1-1.5*IQR))|(df> (Q3+ 1.5*IQR))).any(axis=1)]
```

### Modelling

```python
#Importing necessary libraries



from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=i)
    RFR=RandomForestClassifier()
    RFR.fit(x_train,y_train)
    pred=RFR.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best accuracy is",maxAccu,"at random_state",maxRS)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=maxRS)
```

```python
#Importing all the classification libraries

from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
#Checking accuracy for Random Forest Classifier

RFC=RandomForestClassifier()
RFC.fit(x_train,y_train)
predRFC=RFC.predict(x_test)
print(accuracy_score(y_test,predRFC))
print(confusion_matrix(y_test,predRFC))
print(classification_report(y_test,predRFC))
```

```python
#Accuracy score is 87.40%
```

```python
#Checking accuracy for logistic regression model

LR=LogisticRegression()
LR.fit(x_train,y_train)
predLR=LR.predict(x_test)
print(accuracy_score(y_test,predLR))
print(confusion_matrix(y_test,predLR))
print(classification_report(y_test,predLR))
```

```python
#Accuracy score is 83.22%
```

```python
#Checking accuracy for support vector machine classifier

svc=SVC()
svc.fit(x_train,y_train)
predsvc=svc.predict(x_test)
print(accuracy_score(y_test,predsvc))
print(confusion_matrix(y_test,predsvc))
print(classification_report(y_test,predsvc))
```
