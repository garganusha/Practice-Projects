```python
print("Total duplicate rows are:",df.duplicated().sum())   #THis function is used to check the duplicate rows in the df
```

df.duplicated() returns a boolean Series that indicates whether each row is a duplicate of a previous row. The value True indicates that the row is a duplicate, and False indicates that it is not.

df.duplicated().sum() then calculates the sum of all True values in the boolean Series, which corresponds to the number of duplicated rows in the DataFrame.

Note that the duplicated() method considers all columns by default, but you can specify a subset of columns to consider for duplicates by passing them as an argument. For example, df.duplicated(subset=['Open', 'Close']) will only consider duplicates based on the 'Open' and 'Close' columns.
